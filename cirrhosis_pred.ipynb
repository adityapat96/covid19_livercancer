```python
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression

train_data, holdout_data = train_test_split(s.X, test_size=0.2, random_state=123)

lda = LinearDiscriminantAnalysis()
lr = LogisticRegression()

lda.fit(train_data, s.y[train_data.index]) 
lr.fit(train_data, s.y[train_data.index])
```

```python
lda_preds = lda.predict_proba(holdout_data)
lr_preds = lr.predict_proba(holdout_data)

blended_preds = (lda_preds + lr_preds) / 2

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(s.y[holdout_data.index], blended_preds.argmax(axis=1))
print(f"Blended Accuracy: {accuracy}")
```
